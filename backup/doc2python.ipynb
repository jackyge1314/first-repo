```python
from docx2python import docx2python

# 提取内容（新版直接通过属性访问）
doc = docx2python("example/test.doc")

# 获取正文内容（新版结构）
print("---正文---")
print(doc.body)  # 直接访问body属性
# 获取其他部分
print("\n---页眉---")
print(doc.header)
print("\n---页脚---")
print(doc.footer)
```

```python
# 使用 doc.text 属性
from docx2python import docx2python
text = docx2python("example/test.doc").text
print(text)
```

```python
# 使用字典格式输出
from docx2python import docx2python
import json

doc = docx2python("example/test.doc")

# 手动构建字典
data = {
    "body": doc.body,
    "header": doc.header,
    "footer": doc.footer,
    "text": doc.text  # 所有文本合并后的字符串
}

print(json.dumps(data, indent=2, ensure_ascii=False))
```
